### Exibindo outras Paginas:

```python
import requests
import pandas as pd

api_key = "6a0efdab02802a37b023045239e9c652"
base_url = "https://api.themoviedb.org/3/movie/top_rated"
params = {
    "api_key": api_key,
    "language": "pt-BR",
    "page": 1  # Inicializando na primeira página
}

filmes = []

# Limite de páginas a serem exibidas
limite_paginas = 150

for _ in range(limite_paginas):
    response = requests.get(base_url, params=params)
    data = response.json()
    
    # Adiciona os filmes da página atual
    for movie in data['results']:
        filmes.append({
            'Titulo': movie['title'],
            'Data de lançamento': movie['release_date'],
            'Visão geral': movie['overview'],
            'Votos': movie['vote_count'],
            'Média de votos': movie['vote_average']
        })
    
    # Incrementa para a próxima página
    params["page"] += 1
    if params["page"] > data['total_pages']:
        break  # Sai do loop se não houver mais páginas

# Cria o DataFrame com os filmes
df = pd.DataFrame(filmes)

# Exibe o DataFrame
display(df)
```

### Puxando Filmes do TMDB e Salvando em JSON (CODIGO EM PYTHON)

```python
import requests
import json
import os

# Configurações da API
api_key = "6a0efdab02802a37b023045239e9c652"
base_discover_url = "https://api.themoviedb.org/3/discover/movie"
base_detail_url = "https://api.themoviedb.org/3/movie"
language = "pt-BR"

# Função para buscar filmes básicos da API
def buscar_filmes_basicos(api_key, pagina):
    url = f"{base_discover_url}?api_key={api_key}&language={language}&page={pagina}&primary_release_date.gte=1970-01-01&primary_release_date.lte=2020-12-31&vote_average.gte=8&with_genres=28,12"
    resposta = requests.get(url)
    if resposta.status_code == 200:
        return resposta.json().get("results", [])
    else:
        print(f"Erro ao buscar filmes básicos: {resposta.status_code}")
        return []

# Função para buscar detalhes completos de um filme
def buscar_detalhes_filme(api_key, movie_id):
    url = f"{base_detail_url}/{movie_id}?api_key={api_key}&language={language}"
    resposta = requests.get(url)
    if resposta.status_code == 200:
        return resposta.json()
    else:
        print(f"Erro ao buscar detalhes do filme {movie_id}: {resposta.status_code}")
        return None

# Pasta para salvar os arquivos JSON
pasta_saida = "C:\\Users\\Rafael\\Dev\\Programa_de_Bolsas_Compass_Rafael_Prado\\Sprint 7\\Desafio\\"
os.makedirs(pasta_saida, exist_ok=True)

# Variáveis para coleta de dados
total_filmes = 3000  # 150 páginas * 20 filmes por página
filmes_por_arquivo = 100
pagina_atual = 1
filmes_coletados = []

# Coleta os filmes até atingir o total desejado
while len(filmes_coletados) < total_filmes and pagina_atual <= 150:
    filmes_basicos = buscar_filmes_basicos(api_key, pagina_atual)
    if not filmes_basicos:
        break
    for filme in filmes_basicos:
        detalhes = buscar_detalhes_filme(api_key, filme["id"])
        if detalhes:
            filmes_coletados.append(detalhes)
    pagina_atual += 1

# Limita os filmes ao total desejado
filmes_coletados = filmes_coletados[:total_filmes]

# Divide os filmes em arquivos JSON
for i in range(0, len(filmes_coletados), filmes_por_arquivo):
    parte = i // filmes_por_arquivo + 1
    nome_arquivo = f"tmdb_detailed_data_part_{parte}.json"
    caminho_arquivo = os.path.join(pasta_saida, nome_arquivo)
    with open(caminho_arquivo, "w", encoding="utf-8") as f:
        json.dump(filmes_coletados[i:i + filmes_por_arquivo], f, ensure_ascii=False, indent=4)

print(f"{len(filmes_coletados)} filmes com todos os detalhes salvos em arquivos JSON.")
```

### Concatenar arquivos JSON em DF e passar pra DF

```python
import pandas as pd

# Função para carregar um arquivo JSON
def carregar_arquivo_json(caminho_arquivo):
    return pd.read_json(caminho_arquivo)

# Caminhos para os arquivos JSON (substitua pelos caminhos corretos)
caminho_arquivo1 = 'C:\\Users\\Rafael\\Dev\\Programa_de_Bolsas_Compass_Rafael_Prado\\Sprint 7\\Desafio\\tmdb_data_part_1.json'
caminho_arquivo2 = 'C:\\Users\\Rafael\\Dev\\Programa_de_Bolsas_Compass_Rafael_Prado\\Sprint 7\\Desafio\\tmdb_data_part_2.json'
caminho_arquivo3 = 'C:\\Users\\Rafael\\Dev\\Programa_de_Bolsas_Compass_Rafael_Prado\\Sprint 7\\Desafio\\tmdb_data_part_3.json'


# Carrega os arquivos JSON em DataFrames
df_filmes1 = carregar_arquivo_json(caminho_arquivo1)
df_filmes2 = carregar_arquivo_json(caminho_arquivo2)
df_filmes3 = carregar_arquivo_json(caminho_arquivo3)


# Combina os DataFrames em um único DataFrame
df_filmes = pd.concat([df_filmes1, df_filmes2,df_filmes3], ignore_index=True)

# Exibe as primeiras linhas do DataFrame combinado
print(df_filmes.head())

# Exemplo de organização da tabela: selecionando colunas específicas
colunas_selecionadas = ['title', 'release_date', 'vote_average', 'vote_count']

# Cria um novo DataFrame com as colunas selecionadas
df_filmes_organizado = df_filmes[colunas_selecionadas]

# Exibe as primeiras linhas da tabela organizada
print(df_filmes_organizado.head())

# Salva a tabela organizada em um novo arquivo CSV
df_filmes_organizado.to_csv('filmes_organizados.csv', index=False)

```

### Codigo que Cruza Tabelas

```python
import pandas as pd

# Função para extrair o ano da coluna release_date
def extrair_ano(data):
    return pd.to_datetime(data, errors='coerce').dt.year

# Carrega o CSV gerado anteriormente
df_filmes_organizado = pd.read_csv(r'C:\Users\Rafael\Dev\Programa_de_Bolsas_Compass_Rafael_Prado\Sprint 7\Desafio\filmes_organizados.csv')

# Extrai o ano da coluna release_date
df_filmes_organizado['release_year'] = extrair_ano(df_filmes_organizado['release_date'])

# Carrega o outro CSV que contém a coluna <tituloPrincipal> e <anodelancamento> com o delimitador "|"
df_outro = pd.read_csv(r'C:\Users\Rafael\Dev\Programa_de_Bolsas_Compass_Rafael_Prado\Sprint 6\Desafio\entregaveis\movies.csv', delimiter='|')

# Imprime as colunas do DataFrame para verificação
print("Colunas do df_filmes_organizado:", df_filmes_organizado.columns)
print("Colunas do df_outro antes da renomeação:", df_outro.columns)

# Altera o nome das colunas para corresponder às colunas do primeiro CSV
df_outro.rename(columns={'tituloPincipal': 'title', 'anoLancamento': 'release_year'}, inplace=True)

# Verifica as colunas após a renomeação
print("Colunas do df_outro após a renomeação:", df_outro.columns)

# Remove valores inválidos na coluna release_year
df_outro = df_outro[pd.to_numeric(df_outro['release_year'], errors='coerce').notnull()]

# Converte a coluna release_year para inteiro no segundo DataFrame
df_outro['release_year'] = df_outro['release_year'].astype(int)

# Confere se as colunas 'title' e 'release_year' existem em ambos os DataFrames
if 'title' in df_filmes_organizado.columns and 'release_year' in df_filmes_organizado.columns and 'title' in df_outro.columns and 'release_year' in df_outro.columns:
    # Encontra os filmes em comum nos dois DataFrames
    df_comum = pd.merge(df_filmes_organizado, df_outro, on=['title', 'release_year'])

    # Seleciona apenas as colunas 'title' e 'release_year' para o novo CSV
    df_comum = df_comum[['title', 'release_year']]

    # Remove duplicados
    df_comum.drop_duplicates(subset=['title', 'release_year'], inplace=True)

    # Exibe as primeiras linhas do DataFrame com os filmes em comum e sem duplicados
    print(df_comum.head())

    # Salva o novo DataFrame em um arquivo CSV
    df_comum.to_csv('filmes_comum.csv', index=False)
else:
    print("As colunas 'title' e 'release_year' não foram encontradas em um dos DataFrames.")
```
